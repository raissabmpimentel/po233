# Análise exploratória de dados (Python)

## Importações

```python
# -*- coding: utf-8 -*-

import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

## Matriz de correlação

```python
df_mat = pd.read_csv('dados/student-mat.csv', sep=";")
df_port = pd.read_csv('dados/student-por.csv', sep=";")

df_num_mat = df_mat[['age','absences','G1','G2','G3']]
df_num_port = df_port[['age','absences','G1','G2','G3']]
```

```python
corrMatrix_mat = df_num_mat.corr()
sn.heatmap(corrMatrix_mat, annot=True)
plt.title("Matriz de correlação de atributos quantitativos\n Base matemática")
plt.show()
```

```python
corrMatrix_port = df_num_port.corr()
sn.heatmap(corrMatrix_port, annot=True)
plt.title("Matriz de correlação de atributos quantitativos\n Base português")
plt.show()
```
